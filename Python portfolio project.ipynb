```python
country_respondents = df ['Q11 - Which Country do you live in?'].value_counts().head(10)

country_respondents
```

```python
country_uses_sql= country_grp['Q5 - Favorite Programming Language'].apply(lambda x: x.str.contains('SQL').sum()).head(10)
country_uses_sql
```

```python
sql_df= pd.concat([country_respondents,country_uses_sql], axis='columns', sort=False)
sql_df
```

```python
sql_df.rename(columns={'count': 'Num_Respondents','Q5 - Favorite Programming Language':'SQL_Usage'}, inplace=True)
```

```python
sql_df
```

```python
sql_df['PctKnowsSQL']= (sql_df['SQL_Usage']/sql_df['Num_Respondents'])*100
sql_df
```

```python
sql_df.sort_values(by ='PctKnowsSQL', ascending=False, inplace=True )
```
